```python
t = df.pivot_table('nombre_connexion',
                  aggfunc=np.mean,
                  index = 'freq_connexion',
                  columns='acquis')
```

```python
t.plot(kind='bar', title='Frequence de connexion sur le nombre de connexion')
```
